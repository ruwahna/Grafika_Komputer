```python
# NAMA  : INDAH RUWAHNA ANUGRAHENI
# KELAS : 3IKRB
# NIM   : 240202866

import matplotlib.pyplot as plt
import numpy as np

def dda_line(x1, y1, x2, y2):
    """
    Implementasi Algoritma DDA untuk pembentukan garis
    
    Parameter:
    x1, y1: Koordinat titik awal
    x2, y2: Koordinat titik akhir
    
    Algoritma DDA (Digital Differential Analyzer) adalah algoritma untuk
    menggambar garis lurus pada layar komputer dengan menggunakan perhitungan
    inkremental untuk menentukan piksel-piksel yang harus dinyalakan.
    """
    
    print("=" * 60)
    print("ALGORITMA DDA - PEMBENTUKAN GARIS")
    print("=" * 60)
    
    # ========================================================================
    # LANGKAH 1: Tentukan dua titik yang akan dihubungkan dalam pembentukan garis
    # ========================================================================
    # Ini adalah input dari user - dua titik yang ingin kita hubungkan
    # dengan garis lurus pada layar komputer
    print(f"\n1. Titik awal: ({x1}, {y1})")
    print(f"   Titik akhir: ({x2}, {y2})")
    
    # ========================================================================
    # LANGKAH 2: Tentukan salah satunya sebagai titik awal (x, y) 
    #            dan yang lain sebagai titik akhir (x, y)
    # ========================================================================
    # Kita selalu mulai menggambar dari titik pertama menuju titik kedua
    print(f"\n2. Titik awal yang dipilih: ({x1}, {y1})")
    print(f"   Titik lainnya: ({x2}, {y2})")
    
    # ========================================================================
    # LANGKAH 3: Hitung dx dan dy
    # ========================================================================
    # dx = perubahan koordinat x = x - x (jarak horizontal)
    # dy = perubahan koordinat y = y - y (jarak vertikal)
    # Ini menentukan arah dan panjang garis yang akan digambar
    dx = x2 - x1
    dy = y2 - y1
    print(f"\n3. Hitung dx dan dy:")
    print(f"   dx = x2 - x1 = {x2} - {x1} = {dx}")
    print(f"   dy = y2 - y1 = {y2} - {y1} = {dy}")
    
    # ========================================================================
    # LANGKAH 4: Tentukan step, dengan ketentuan berikut:
    # ========================================================================
    # - Jika |dx| > |dy|, maka step = |dx|  (garis lebih horizontal)
    # - Jika tidak, maka step = |dy|        (garis lebih vertikal)
    # 
    # Step menentukan berapa banyak piksel yang akan digambar.
    # Kita memilih yang lebih besar agar tidak ada "gap" (celah) pada garis.
    if abs(dx) > abs(dy):
        step = abs(dx)
        print(f"\n4. Karena |dx| = {abs(dx)} > |dy| = {abs(dy)}")
        print(f"   Maka step = |dx| = {step}")
    else:
        step = abs(dy)
        print(f"\n4. Karena |dx| = {abs(dx)} <= |dy| = {abs(dy)}")
        print(f"   Maka step = |dy| = {step}")
    
    # ========================================================================
    # LANGKAH 5: Hitung penambahan koordinat piksel dengan persamaan:
    # ========================================================================
    # x_inc = dx / step  (penambahan x setiap iterasi)
    # y_inc = dy / step  (penambahan y setiap iterasi)
    # 
    # Ini adalah "increment" atau langkah kecil yang diambil setiap kali
    # kita menggambar piksel berikutnya. Nilai ini bisa desimal (floating point).
    if step == 0:
        x_inc = 0
        y_inc = 0
    else:
        x_inc = dx / step
        y_inc = dy / step
    
    print(f"\n5. Hitung penambahan koordinat piksel:")
    print(f"   x_inc = dx / step = {dx} / {step} = {x_inc:.4f}")
    print(f"   y_inc = dy / step = {dy} / {step} = {y_inc:.4f}")
    
    # ========================================================================
    # LANGKAH 6: Koordinat selanjutnya dihitung dengan:
    # ========================================================================
    # x = x + x_inc
    # y = y + y_inc
    # 
    # Kita mulai dari titik awal (x, y), kemudian terus menambahkan
    # x_inc dan y_inc untuk mendapatkan posisi piksel berikutnya
    x = x1
    y = y1
    print(f"\n6. Koordinat awal:")
    print(f"   x = x1 + x_inc = {x1}")
    print(f"   y = y1 + y_inc = {y1}")
    
    # Menyimpan koordinat piksel untuk plotting nanti
    x_coords = []
    y_coords = []
    
    # ========================================================================
    # LANGKAH 7: Lakukan pembulatan u = Round(x), v = Round(y)
    #            kemudian plot piksel (u, v) pada layar
    # ========================================================================
    # Karena layar komputer hanya bisa menampilkan piksel pada koordinat
    # integer (bilangan bulat), kita harus membulatkan nilai x dan y
    # yang mungkin berupa desimal
    print(f"\n7. Pembulatan koordinat dan plotting:")
    print(f"   {'Iterasi':<10} {'x':<15} {'y':<15} {'Round(x)':<12} {'Round(y)':<12}")
    print("   " + "-" * 70)
    
    # ========================================================================
    # LANGKAH 8: Ulangi point 6 dan 7 untuk menentukan posisi piksel
    #            berikutnya sampai x = x dan y = y
    # ========================================================================
    # Loop ini akan berjalan sebanyak (step + 1) kali untuk menggambar
    # semua piksel dari titik awal sampai titik akhir
    for i in range(int(step) + 1):
        # LANGKAH 7: Pembulatan nilai x dan y ke integer terdekat
        x_plot = round(x)
        y_plot = round(y)
        
        # Simpan koordinat piksel yang sudah dibulatkan
        x_coords.append(x_plot)
        y_coords.append(y_plot)
        
        # Tampilkan proses perhitungan setiap iterasi
        print(f"   {i:<10} {x:<15.4f} {y:<15.4f} {x_plot:<12} {y_plot:<12}")
        
        # LANGKAH 6 & 8: Update koordinat untuk iterasi berikutnya
        # Tambahkan x_inc ke x, dan y_inc ke y
        x = x + x_inc
        y = y + y_inc
    
    print(f"\n8. Proses selesai! Total {len(x_coords)} piksel diplot")
    print("=" * 60)
    
    return x_coords, y_coords


def plot_line(x1, y1, x2, y2, x_coords, y_coords):
    """
    Visualisasi hasil algoritma DDA
    
    Fungsi ini membuat dua plot:
    1. Perbandingan antara garis ideal (matematika) dengan piksel DDA
    2. Grid piksel yang menunjukkan urutan penggambaran
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # ========================================================================
    # PLOT 1: Perbandingan Garis Ideal vs Piksel DDA
    # ========================================================================
    # Garis biru: garis matematika yang sempurna (kontinyu)
    # Titik merah: piksel-piksel yang benar-benar digambar di layar (diskrit)
    ax1.plot([x1, x2], [y1, y2], 'b-', linewidth=2, label='Garis Ideal', alpha=0.5)
    ax1.plot(x_coords, y_coords, 'ro-', markersize=8, label='Piksel DDA', linewidth=1)
    ax1.grid(True, alpha=0.3)
    ax1.set_xlabel('X', fontsize=12)
    ax1.set_ylabel('Y', fontsize=12)
    ax1.set_title('DDA Line Algorithm - Perbandingan', fontsize=14, fontweight='bold')
    ax1.legend()
    ax1.axis('equal')
    
    # ========================================================================
    # PLOT 2: Grid Piksel dengan Urutan Penggambaran
    # ========================================================================
    # Setiap kotak mewakili satu piksel di layar
    # Angka di dalam kotak menunjukkan urutan piksel digambar (0, 1, 2, ...)
    # Warna gradient menunjukkan progres dari awal (gelap) ke akhir (terang)
    ax2.scatter(x_coords, y_coords, c=range(len(x_coords)), cmap='viridis', 
                s=200, marker='s', edgecolors='black', linewidth=2)
    
    # Tambahkan nomor urutan pada setiap piksel
    for i, (x, y) in enumerate(zip(x_coords, y_coords)):
        ax2.text(x, y, str(i), ha='center', va='center', 
                color='white', fontweight='bold', fontsize=10)
    
    ax2.grid(True, alpha=0.3)
    ax2.set_xlabel('X', fontsize=12)
    ax2.set_ylabel('Y', fontsize=12)
    ax2.set_title('DDA Line Algorithm - Urutan Piksel', fontsize=14, fontweight='bold')
    ax2.axis('equal')
    
    plt.tight_layout()
    plt.show()


# ============================================================================
# CONTOH PENGGUNAAN ALGORITMA DDA
# ============================================================================
# Di bawah ini adalah beberapa contoh penerapan algoritma DDA
# untuk berbagai jenis garis
if __name__ == "__main__":
    # ------------------------------------------------------------------------
    # Contoh 1: Garis dengan kemiringan landai (|dx| > |dy|)
    # ------------------------------------------------------------------------
    # Garis lebih horizontal, step ditentukan oleh dx
    print("\n### CONTOH 1: Garis Landai ###")
    x1, y1 = 2, 3
    x2, y2 = 10, 6
    x_coords, y_coords = dda_line(x1, y1, x2, y2)
    plot_line(x1, y1, x2, y2, x_coords, y_coords)
    
    # ------------------------------------------------------------------------
    # Contoh 2: Garis dengan kemiringan curam (|dy| > |dx|)
    # ------------------------------------------------------------------------
    # Garis lebih vertikal, step ditentukan oleh dy
    print("\n\n### CONTOH 2: Garis Curam ###")
    x1, y1 = 1, 1
    x2, y2 = 5, 9
    x_coords, y_coords = dda_line(x1, y1, x2, y2)
    plot_line(x1, y1, x2, y2, x_coords, y_coords)
    
    # ------------------------------------------------------------------------
    # Contoh 3: Garis horizontal (dy = 0)
    # ------------------------------------------------------------------------
    # Garis sejajar sumbu X, hanya koordinat x yang berubah
    print("\n\n### CONTOH 3: Garis Horizontal ###")
    x1, y1 = 1, 5
    x2, y2 = 8, 5
    x_coords, y_coords = dda_line(x1, y1, x2, y2)
    plot_line(x1, y1, x2, y2, x_coords, y_coords)
```
