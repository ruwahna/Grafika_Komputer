```python
import matplotlib.pyplot as plt

# --- 1. Inisialisasi Input ---

# Daftar titik koordinat segitiga P = {(x1, y1), (x2, y2), (x3, y3)}
# Ini adalah titik-titik segitiga ASLI (A, B, C)
titik_awal = [
    (2, 2),  # Titik A
    (6, 2),  # Titik B
    (4, 6)   # Titik C
]

# Faktor Skala
sx = 2
sy = 2

# Wadah untuk menyimpan hasil (titik baru P')
titik_baru = []

print("--- Proses Perhitungan Scaling ---")

# --- 2. Proses Looping (Perhitungan sesuai algoritma) ---
for i, titik in enumerate(titik_awal):
    x, y = titik
    
    # Hitung koordinat baru (Rumus: x' = x * sx, y' = y * sy)
    x_baru = x * sx
    y_baru = y * sy
    
    # Simpan titik baru
    titik_baru.append((x_baru, y_baru))
    
    # Cetak proses perhitungan ke konsol
    nama_titik = chr(65 + i) # Untuk memberi nama A, B, C
    print(f"Titik {nama_titik} ({x}, {y}) -> {nama_titik}' ({x}*{sx}, {y}*{sy}) = ({x_baru}, {y_baru})")

# --- 3. Output Hasil Akhir (Cetak ke konsol) ---
print("\n--- Hasil Akhir (Koordinat Baru) ---")
print(f"Daftar titik awal: {titik_awal}")
print(f"Daftar titik setelah scaling: {titik_baru}")

# --- 4. Visualisasi Grafik (Menggunakan Matplotlib) ---

# Mengubah format list of tuples menjadi dua list terpisah (x_coords, y_coords)
# dan menambahkan titik awal lagi di akhir untuk menutup bentuk segitiga.
x_coords_awal, y_coords_awal = zip(*titik_awal + [titik_awal[0]])
x_coords_akhir, y_coords_akhir = zip(*titik_baru + [titik_baru[0]])

plt.figure(figsize=(8, 8)) # Ukuran plot
plt.plot(x_coords_awal, y_coords_awal, 'ro-', label='Segitiga Awal') # 'ro-' = red, circle marker, line
plt.plot(x_coords_akhir, y_coords_akhir, 'bs-', label='Segitiga Setelah Scaling') # 'bs-' = blue, square marker, line

# Menambahkan label pada setiap titik untuk segitiga awal
for i, (x, y) in enumerate(titik_awal):
    plt.text(x, y, f' {chr(65+i)} ({x},{y})', color='red', fontsize=10, ha='right')

# Menambahkan label pada setiap titik untuk segitiga setelah scaling
for i, (x, y) in enumerate(titik_baru):
    plt.text(x, y, f' {chr(65+i)}\' ({x},{y})', color='blue', fontsize=10, ha='left')


plt.title('Transformasi Scaling pada Segitiga')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.axhline(0, color='gray', linewidth=0.5) # Garis sumbu X di 0
plt.axvline(0, color='gray', linewidth=0.5) # Garis sumbu Y di 0
plt.grid(True) # Menampilkan grid
plt.axis('equal') # Memastikan skala sumbu X dan Y sama
plt.legend() # Menampilkan legenda
plt.show() # Menampilkan plot
```
