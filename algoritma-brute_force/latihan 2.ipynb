```python
# NAMA : INDAH RUWAHNA ANUGRAHENI
# KELAS : 3IKRB 
# NIM : 240202866


import matplotlib.pyplot as plt

def bruteforce_algorithm(x1,y1,x2,y2):
    # buat list 
    points = []

    # perhitungan dx dan dy 
    dx = x2 - x1
    dy = y2 - y1

    # jika dx lebih besar dari dy
    if abs(dx) >= abs(dy):
        # hitung kemiringan slope (m)        
        m = dy / dx if dx != 0 else 0 # agar tidak dibagi 0  
        # perulangan 
    for x in range (x1, x2 + 1) if x1 < x2 else range (x1, x2 - 1, -1):
        # y = mx + c
        y = y1 + (m * (x - x1))
        points.append((x, y))
# jika dy lebih besar dari dx
    else:
        # hitung kemiringan slope (m)
        m_invers = dx / dy if dy != 0 else 0 # agar tidak dibagi 0
        for y in range (y1, y2 + 1) if x1 < x2 else range (y1, y2 - 1, -1):
            x1 = x1 + int(m_invers * (y - y1))
            points.append((x1, y))
        #  mengembalikan nilai  list berdasarkan titik yang telah ditentukan
            return points
        
def plot_brute_force(x1, y1, x2, y2):
    kordinat = bruteforce_algorithm(x1, y1, x2, y2)
    
    x_coords = [point[0] for point in kordinat]
    y_coords = [point[1] for point in kordinat]
    
    plt.figure(figsize=(6,6))
    plt.plot(x_coords, y_coords, marker='o', color='blue', markersize=5, linestyle='-', linewidth=1)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title("ALGORITMA BRUTEFORCE")
    plt.grid(True)
    plt.show()
    
    return x_coords, y_coords

# hasil
x1, y1 = 1, 2
x2, y2 = 5, 5

grafik = plot_brute_force(x1, y1, x2, y2)
bf = bruteforce_algorithm(x1, y1, x2, y2)
print(bf)
print(grafik)
```
