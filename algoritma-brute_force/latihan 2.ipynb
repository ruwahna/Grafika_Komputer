```python
import matplotlib.pyplot as plt
import math

# =======================================================
# 1. Input Titik (Ditetapkan Langsung)
# =======================================================
# Koordinat yang diuji: (2, 2) ke (10, 5)
x1, y1, x2, y2 = 2, 2, 10, 5 

print("=============================================")
print(f"  Program Brute Force: A({x1}, {y1}) ke B({x2}, {y2})")
print("=============================================")


# =======================================================
# 2. Algoritma Brute Force (Fungsi)
# =======================================================
def draw_line_bruteforce(x1, y1, x2, y2):
    """Menghitung piksel garis menggunakan algoritma Brute Force."""
    
    pixel_coordinates = []

    # Pertukaran Titik: Pastikan iterasi dari kiri ke kanan (x1 <= x2)
    if x1 > x2:
        x1, x2 = x2, x1
        y1, y2 = y2, y1
        
    # --- Penanganan Kasus Khusus (Horizontal/Vertikal) ---
    if y1 == y2: 
        for x in range(x1, x2 + 1):
            pixel_coordinates.append((x, y1))
            
    elif x1 == x2: 
        if y1 > y2:
            y1, y2 = y2, y1
        for y in range(y1, y2 + 1):
            pixel_coordinates.append((x1, y))

    # --- Kasus Umum ---
    else:
        M = (y2 - y1) / (x2 - x1) 
        
        for x in range(x1, x2 + 1):
            y_float = M * (x - x1) + y1
            y_int = round(y_float) 
            
            pixel_coordinates.append((x, y_int))
            
    return pixel_coordinates 


# =======================================================
# Panggilan Fungsi Utama (Melanjutkan Poin 2)
# =======================================================
hasil_koordinat = draw_line_bruteforce(x1, y1, x2, y2)


# =======================================================
# 3. Menampilkan Daftar Koordinat Titik Piksel yang Terbentuk
# =======================================================
print("  Daftar Koordinat Piksel yang Terbentuk:")
print("---------------------------------------------")
if hasil_koordinat:
    for coord in hasil_koordinat:
        print(coord)
else:
    print("Tidak ada koordinat yang dihasilkan.")
print("=============================================")


# =======================================================
# 4. Menampilkan Hasil Garis Menggunakan Matplotlib
# (Garis Bersambung dan Marker Titik)
# =======================================================
if hasil_koordinat:
    # Pisahkan koordinat X dan Y
    x_coords = [c[0] for c in hasil_koordinat]
    y_coords = [c[1] for c in hasil_koordinat]

    # Menggambar Garis Lurus (Persis seperti gambar yang Anda kirimkan)
    plt.plot(
        x_coords, 
        y_coords, 
        color='blue', 
        linestyle='-', 
        linewidth=1.5,
        marker='o',       # Menambahkan Marker (Titik)
        markersize=8,     # Ukuran Titik
        markerfacecolor='blue',
        label='Garis Hasil Algoritma'
    )
    
    # Pengaturan Visualisasi
    plt.title("ALGORITMA BRUTEFORCE") # Sesuai dengan gambar Anda
    plt.xlabel("X Axis")
    plt.ylabel("Y Axis")
    plt.grid(True)
    
    # PENTING: Atur batas sumbu agar plot terlihat jelas
    all_coords = [x1, y1, x2, y2]
    if all_coords:
        plot_limit = max(abs(val) for val in all_coords) + 5
        plt.xlim(min(x_coords) - 1, max(x_coords) + 1)
        plt.ylim(min(y_coords) - 1, max(y_coords) + 1)
        
    plt.gca().set_aspect('equal', adjustable='box') 
    
    plt.show() # Tampilkan gambar
```
